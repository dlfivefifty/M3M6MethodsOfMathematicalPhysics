## Integrals over the real line

Integrals on the real line are always to viewed as improper integrals:

$$ \int_{-\infty}^\infty f(x) dx = \int_0^\infty f(x)dx +\int_{-\infty}^0 f(x)dx = \lim_{b \rightarrow \infty} \int_0^b f(x) dx + \lim_{a \rightarrow -\infty} \int_a^0 f(x) dx.$$

### Residue theorem on the real line

The real line doesn't have an _inside_ and _outside_, rather an _above_ and _below_, or _left_ and _right_. Thus we get the following two versions of the Residue theorem:

**Definition (Upper/lower half plane)** Denote the upper/lower half plane by
$$
{\mathbb H}^+ = \{z : \Re z > 0 \}  \\
{\mathbb H}^- = \{z : \Re z < 0 \} 
$$
The closure is denoted
$$
\bar{\mathbb H}^+ = {\mathbb H}^+ \cup {\mathbb R} \cup \{\infty\}  \\
\bar{\mathbb H}^- = {\mathbb H}^- \cup {\mathbb R} \cup \{\infty\}
$$

**Theorem (Residue theorem on the real line)** Suppose $f : \bar {\mathbb H}^+ \backslash \{z_1,\ldots,z_r \} \rightarrow {\mathbb C}$ is holomorphic in ${\mathbb H}^+ \backslash \{z_1,\ldots,z_r \}$, where $\Re z_k > 0$, and  $\lim_{\epsilon \rightarrow 0} f(x + i \epsilon) = f(x)$ converges uniformly.  If 
$$ \lim_{z \rightarrow \infty} z f(z) = 0
$$ uniformly for $z \in \bar {\mathbb H}^+$, then
$$
\int_{-\infty}^\infty f(x) = 2 \pi i \sum_{k=1}^r {\underset{z = z_k}{\rm Res}} \, f(z)
$$
Similarly, if the equivalent conditions hold in the lower half plane for $f : \bar{\mathbb H}^- \backslash \{z_1,\ldots,z_r \} \rightarrow {\mathbb C}$ then
$$
\int_{-\infty}^\infty f(x) = -2 \pi i \sum_{k=1}^r {\underset{z = z_k}{\rm Res}} \, f(z)
$$


Examples:

```julia
f = x -> x^2/(x^4+1)
phaseplot(f, (-3,3),(-2,2))
```

```julia
sum(Fun(f, -Inf .. Inf))
```

```julia
z,z,z,z = exp(im*/4), exp(3im*/4), exp(5im*/4), exp(7im*/4)
f(z),f(z)
```

```julia
res = z^2 / ((z - z)*(z - z)*(z - z) )
res = z^2 / ((z - z)*(z - z)*(z - z) )

2*im*(res + res)
```

```julia
res = z^2 / ((z - z)*(z - z)*(z - z) )
res = z^2 / ((z - z)*(z - z)*(z - z) )

-2*im*(res + res)
```

An immediate consequence of this result is Cauchy's integral formula on the real line:

**Theorem (Cauchy's integral formula on the real line)** Suppose $f : \bar {\mathbb H}^+  \rightarrow {\mathbb C}$ is holomorphic in ${\mathbb H}^+ $, and  $\lim_{\epsilon \rightarrow 0} f(x + i \epsilon) = f(x)$ converges uniformly.  If 
$$ 
\lim_{z \rightarrow \infty} f(z) = 0
$$ 
uniformly for $z \in \bar {\mathbb H}^+$, then
$$
f(z) = {1 \over 2 \pi i} \int_{-\infty}^\infty {f(x) \over x - z} dx
$$
for all  $z \in {\mathbb H}^+$.

```julia
f = x -> x^2/(x+im)^3
z = 2.0+2.0im
sum(Fun(x-> f(x)/(x - z), -Inf .. Inf))/(2*im) - f(z)
```

```julia
z = -2-im
sum(Fun(x-> f(x)/(x - z), -500 .. 500))/(2*im)
```

### Jordan's lemma

The case of calculating
$$
    \int_{-\infty}^\infty e^{i \omega x} g(x) dx
$$ 
is important because it is the Fourier transform of $g(x)$. Provided $g$ is defined in the upper half plane and $\omega > 0$, $f(z) = e^{i \omega z} g(z)$  has exponential decay:

```julia
phaseplot(z -> exp(2im*z) / (z - im), (-4,4), (-4,4))
```

We can use this to get sharper results than ML inequality:

**Lemma (Jordan)** If $g(z)$ is continuous in on the half circle $C_R = \{ R e^{i \theta} : 0 \leq \theta \leq \pi \}$  then
$$\left| \int_{C_R} g(z) e^{i \omega z} dz \right| \leq {\pi \over \omega} M$$
where $M = \sup_{z \in C_R} |g(z)|$. 

**Sketch of proof** We have
$$\left| \int_{C_R} g(z) e^{i \omega z} dz \right|  \leq   R \int_0^\pi \left|g(R e^{i \theta}) e^{i \omega R e^{i \theta}}e^{i \theta}\right| d\theta 
\leq MR \int_0^\pi e^{- \omega R\sin \theta } d\theta 
= 2MR \int_0^{\pi\over 2} e^{- \omega R\sin \theta } d\theta 
$$
But we have $\sin \theta \geq {2 \theta \over \pi}$:

```julia
 = linspace(0,/2,100)
plot(, sin.(); label="sin t")
plot!(, 2/; label = "2t / pi")
```

Hence 
$$\left| \int_{C_R} g(z) e^{i \omega z} dz \right|  \leq  2MR \int_0^{\pi\over 2} e^{- {2\omega R\theta \over \pi} } d\theta = {\pi \over \omega} (1 - e^{-\omega R}) M \leq {\pi M \over \omega}.
$$

## Application: Calculating Fourier integrals of weakly decaying functions

Why is this useful? We can use it to apply Residue theorem to We already know $O(z^{-2})$ decay gives us Residue theorem. And if we only have $z^{-1}$ decay our integral does not converge absolutely:

```julia
f = x -> exp(im*x)*x/(x^2+1)

sum(abs.(Fun(f, 0 .. 2000)))
```

However, it does converge conditionally:

```julia
f = x -> exp(im*x)*x/(x^2+1)

sum(Fun(f, -30000 .. 30000))
```

Thus we can construct a Residue theorem when $g(z) \rightarrow 0$ and is analytic in the upper-half plane.

```julia
2*im*exp(-1)*im/(im+im)
```

## Functions with branch cuts

We now discuss functions with branch cuts

1. $\log z$ with a cut on $(-\infty,0]$
2. $z^\alpha$ with a cut on $(-\infty,0]$
3. $\sqrt{z-1}\sqrt{z+1}$ with a cut on $[-1,1]$
4. $\rm{sign}(|z|-1)$ with a cut on the unit circle.

This is a step towards a Cauchy integral theorem on cuts, for recovering a holomorphic function from its behaviour on a cut.

We begin with $\log z$. 

**Definition (Complex Logarithm)**
$$\log z = \int_1^z {1 \over \zeta} d\zeta$$
where the integral is understood to be on a straight line segment, that is
$$\log z = \int_{\gamma_z} {1 \over \zeta} d\zeta$$
where
$\gamma_z(t) = 1 + (z-1)t$ for $0 \leq t \leq 1$.

```julia
z = -2 + im
phaseplot( -> 1/, (-3,3), (-3,3))
plot!(Segment(1,conj(z)); color=:black, label="contour")
```

This is well-defined apart from $z \in (-\infty,0]$, where there is a pole on the contour:

```julia
log(-1)
```

```julia
phaseplot(z -> log(z), (-3,3), (-3,3))
```

It does exist in the limit, though:

```julia
log(-1+0.0im), log(-1-0.0im)
```

By deformation, the value of the integrals is independent of the path:

```julia
log( exp(im*0.9))
```

```julia
sum(Fun(z -> 1/z, Arc(0.,1.,(0.,0.9))))
```

```julia
phaseplot( -> 1/, (-3,3), (-3,3))
plot!(Segment(1.0, exp(0.9*im)); ratio=1.0, color=:black, label="original contour")
plot!(Arc(0.,1.,(0.,0.9)); ratio=1.0, color=:grey, label="deformed contour")
```

This works all the way to the negative real axis:

```julia
log(-1+0im), log(-1.0 -0im)
```

```julia
sum(Fun( -> 1/, Arc(0.,1.,(0.,)))),
sum(Fun( -> 1/, Arc(0.,1.,(0.,-))))
```

Combining the two contours we have:

```julia
log(-1.0+0im) - log(-1.0-0im),

sum(Fun( -> 1/, Circle()))
```

### Algebraic powers
We define an algebraic power as:

**Definition (algebraic power)** 
$$z^\alpha = e^{\alpha \log z}$$

Note, for example, when $\alpha = 1/2$, $z^{1/2}$ is only one solution to $y^2 = z$.

```julia
 = 0.9
phaseplot(z -> z^, (-3,3), (-3,3))
```

This has a branch cut on $(-\infty,0]$ on which it has the jump:

$${x_+^\alpha \over x_-^\alpha} = e^{\alpha (\log_+ x - \log_- x)} = e^{2 \pi i \alpha}$$

In particular,

$$\sqrt{x}_+ = -\sqrt{x}_- = i \sqrt{|x|}$$

```julia
sqrt(-1-0.0im) , -sqrt(-1 + 0.0im)
```

Let's look at another example: $\varphi(z) = \sqrt{z-1}\sqrt{z+1}$. 

```julia
 = z -> sqrt(z-1)*sqrt(z+1)
phaseplot(, (-3,3), (-3,3))
```

For $x > 0$ it's holomorphic. For $-1 < x < 1$ we have

$$\varphi_+(x) = \sqrt{x-1}_+ \sqrt{x+1} = - \sqrt{x-1}_- \sqrt{x+1} = -\varphi_-(x)$$

For $x < -1$ we actually have continuity:

$$\varphi_+(x) = \sqrt{x-1}_+ \sqrt{x+1}_+ = (- \sqrt{x-1}_-)(- \sqrt{x+1}_-) = \varphi_-(x)$$

```julia
 = z -> sqrt(z-1)sqrt(z+1)

(0.1+0im), (0.1-0im)
```

```julia
(-2.0+0im), (-2.0-0im)
```


But continuity _implies_ analyticity:

**Theorem (continuity on a curve implies analyticity)** Let $D$ be a domain and $\gamma \subset D$ a contour, where $\gamma(t)$ is defined on $[a,b]$. If $f$ is analytic in $D \backslash \Gamma$, continuous on $\gamma \circ (a,b)$, and has weaker than pole singularities at $\gamma(a)$ and $\gamma(b)$, then $f$ is analytic in all of $D$.

**Sketch of Proof** Use Cauchy integral formula.
